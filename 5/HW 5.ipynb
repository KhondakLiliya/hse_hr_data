Домашнее задание про критерий Шапиро-Уилка

```python
import pandas as pd
from scipy.stats import shapiro
from matplotlib import pyplot as plt
```

```python
df_hr = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

H0 для критерия Шапиро-Уилка звучит как "величина Х распределена нормально".

```python
print("Shapiro-Wilk normality test, W-statistic: %f, p-value: %f" % shapiro(df_hr['MonthlyIncome']))
```

p-value должно быть больше 0,05, чтобы гипотеза о происхождении данных из нормального распределения была принята.
В данном случае мы видим, что полученное значение < 0,05, значит гипотеза не принимается.

```python
plt.hist(df_hr['MonthlyIncome'])
plt.show()
```

Из графика видно, то выбранная величина не имеет нормального распределения
